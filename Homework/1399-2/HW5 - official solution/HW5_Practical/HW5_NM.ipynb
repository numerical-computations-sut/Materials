#Get the Plot

```python
import pandas as pd
import matplotlib.pyplot as plt

path = "data.xlsx"              #Enter the path to "data.xlsx"

df = pd.read_excel(path)

X       = df['x_n'].to_list()
eu      = df['Euler'].to_list()
meu     = df['Modified Euler'].to_list()
rk4     = df['RK4'].to_list()
actual  = df['Actual'].to_list()

plt.plot(X, eu,  color='r',     label='Euler')
plt.plot(X, meu, color='g',     label='Modified Euler')
plt.plot(X, rk4, color='b',     label='RK4')
plt.plot(X, actual, color='black', label='Actual Values')


plt.xlabel("X")
plt.ylabel("Y")
plt.title("Comparisson of Actual values and Numerical Methods")
plt.legend()
  
plt.show()
```
